```python
import gzip
import json
import pandas as pd
from datetime import datetime
```

```python
with gzip.open('steam_reviews.json.gz', 'r') as f:
     reviews = []
     for l in f:
         reviews.append(eval(l))
len(reviews)
```

```python
with gzip.open('steam_games.json.gz', 'r') as f:
    games = []
    for l in f:
        games.append(eval(l))
len(games)
```

```python
begin_date = '2017-01-01'
```

```python
len(review_df[review_df['date'] > begin_date])
```

```python
review_df_filtered_by_date = review_df[review_df['date'] > begin_date]
```

```python
user_count = review_df_filtered_by_date['user_id'].value_counts()
valid_users = user_count[user_count >= 5].index
len(valid_users)
```

```python
len(review_df_filtered_by_date['user_id'].unique())
```

```python
reviews_df_filtered = review_df_filtered_by_date[review_df_filtered_by_date['user_id'].isin(valid_users)]
len(reviews_df_filtered)
```

```python
valid_items = reviews_df_filtered['product_id'].unique()
len(valid_items)
```

```python
user_mapping = {uid: idx + 1 for idx, uid in enumerate(valid_users)}
item_mapping = {iid: idx + 1 for idx, iid in enumerate(valid_items)}
```

```python
games_filtered = games_df[games_df['id'].isin(valid_items)]
len(games_filtered)
```

```python
games_filtered['item_id'] = games_filtered['id'].apply(lambda x: item_mapping[x])
games_filtered.head()
```

```python
reviews_df_filtered_mapped = reviews_df_filtered.copy()
reviews_df_filtered_mapped['item_id'] = reviews_df_filtered_mapped['product_id'].apply(lambda x: item_mapping[x])
reviews_df_filtered_mapped['user_id'] = reviews_df_filtered_mapped['user_id'].apply(lambda x: user_mapping[x])
reviews_df_filtered_mapped.head()
```

```python
games_filtered['name'] = games_filtered['app_name']
games_filtered.head()
```

```python
def date2timestamp(date_str: str) -> int:
    """
    毫秒级时间戳 
    """
    date_obj = datetime.strptime(date_str, "%Y-%m-%d")
    timestamp_seconds = datetime.timestamp(date_obj)
    return timestamp_seconds * 1000
```

```python
reviews_df_filtered_mapped['timestamp'] = reviews_df_filtered_mapped['date'].apply(lambda x: date2timestamp(x))
reviews_df_filtered_mapped
```

```python
review_kept_columns = ['user_id', 'item_id', 'timestamp']
games_kept_columns = ['item_id', 'name', 'sentiment', 'genres', 'tags', 'specs']
```

```python
reviews_df_filtered_mapped[review_kept_columns].to_csv("steam_interactions.csv", index=False)
games_filtered[games_kept_columns].to_csv("steam_games.csv", index=False)
```
