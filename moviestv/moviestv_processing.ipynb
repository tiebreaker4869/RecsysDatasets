# Amazon Reviews Movies&TV

```python
import gzip
import json
import pandas as pd
from datetime import datetime
```

```python
# read csv.gz file
def read_csv_gz(file_path):
    with gzip.open(file_path, 'rt', encoding='utf-8') as f:
        df = pd.read_csv(f)
    return df
```

```python
reviews_df = read_csv_gz('Movies_and_TV.csv.gz')
reviews_df.head()
```

```python
with gzip.open('meta_Movies_and_TV.jsonl.gz', 'r') as f:
    meta = []
    for l in f:
        meta.append(json.loads(l))
meta[0]
```

```python
movies_df = pd.DataFrame(meta)
movies_df.head()
```

```python
movies_df.drop_duplicates(subset=['parent_asin'], inplace=True)
```

```python
len(movies_df)
```

```python
movies_df.dropna(subset=['title', 'description', 'categories'], inplace=True)
len(movies_df)
```

```python
all_movies = set(movies_df['parent_asin'].unique())
all_movies
```

```python
reviews_df = reviews_df[reviews_df['parent_asin'].isin(all_movies)]
len(reviews_df)
```

```python
len(reviews_df)
```

```python
len(movies_df)
```

```python
reviews_df.sort_values(by='timestamp', inplace=True, ascending=False)
reviews_df.head()
```

```python
reviews_df['review_date'] = reviews_df['timestamp'].apply(lambda x: datetime.fromtimestamp(x/1000).strftime('%Y-%m-%d'))
reviews_df.head()
```

```python
len(reviews_df[reviews_df['review_date'] > '2020-01-01'])
```

```python
review_df_filtered_by_date = reviews_df[reviews_df['review_date'] > '2020-01-01']
review_df_filtered_by_date.head()
```

```python
user_count = review_df_filtered_by_date['user_id'].value_counts()
valid_users = user_count[user_count >= 5].index
len(valid_users)
```

```python
len(review_df_filtered_by_date['user_id'].unique())
```

```python
reviews_df_filtered = review_df_filtered_by_date[review_df_filtered_by_date['user_id'].isin(valid_users)]
len(reviews_df_filtered)
```

```python
valid_items = reviews_df_filtered['parent_asin'].unique()
len(valid_items)
```

```python
user_mapping = {uid: idx + 1 for idx, uid in enumerate(valid_users)}
item_mapping = {iid: idx + 1 for idx, iid in enumerate(valid_items)}
```

```python
movies_filtered = movies_df[movies_df['parent_asin'].isin(valid_items)]
len(movies_filtered)
```

```python
movies_filtered['item_id'] = movies_filtered['parent_asin'].map(item_mapping)
movies_filtered.head()
```

```python
reviews_df_filtered_mapped = reviews_df_filtered.copy()
reviews_df_filtered_mapped['user_id'] = reviews_df_filtered_mapped['user_id'].map(user_mapping)
reviews_df_filtered_mapped['parent_asin'] = reviews_df_filtered_mapped['parent_asin'].map(item_mapping)
reviews_df_filtered_mapped['item_id'] = reviews_df_filtered_mapped['parent_asin']
reviews_df_filtered_mapped.head()
```

```python
movies_filtered.head()
```

```python
review_kept_columns = ['user_id', 'item_id', 'timestamp', 'rating']
movies_kept_columns = ['item_id', 'title', 'description', 'categories']
```

```python
movies_filtered[movies_kept_columns].isna().sum()
```

```python
reviews_df_filtered_mapped[review_kept_columns].to_csv('moviestv_interactions.csv', index=False)
movies_filtered[movies_kept_columns].to_csv('moviestv.csv', index=False)
```
