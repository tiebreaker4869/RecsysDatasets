```python
begin_date = '2020-01-01 00:00:00'
```

```python
len(reviews_df[reviews_df['date'] > begin_date])
```

```python
reviews_df_filtered_date = reviews_df[reviews_df['date'] > begin_date]
reviews_df_filtered_date.head()
```

```python
len(reviews_df_filtered_date)
```

```python
user_count = reviews_df_filtered_date['user_id'].value_counts()
valid_users = user_count[user_count >= 5].index
reviews_df_filtered = reviews_df_filtered_date[reviews_df_filtered_date['user_id'].isin(valid_users)]
len(reviews_df_filtered)
```

```python
len(reviews_df_filtered['user_id'].unique())
```

```python
user_count_new = reviews_df_filtered['user_id'].value_counts()
long_users = user_count_new[user_count_new >= 20].index
len(long_users)
```

```python
valid_items = reviews_df_filtered['business_id'].unique()
len(valid_items)
```

```python
user_mapping = {uid: idx + 1 for idx, uid in enumerate(valid_users)}
user_mapping
```

```python
valid_items[0]
```

```python
item_mapping = {iid: idx + 1 for idx, iid in enumerate(valid_items)}
```

```python
item_mapping
```

```python
business_filtered = business_df[business_df['business_id'].isin(valid_items)]
business_filtered['item_id'] = business_filtered['business_id'].apply(lambda x: item_mapping[x])
business_filtered.head()
```

```python
reviews_df_filtered.head()
```

```python
reviews_df_filtered_mapped = reviews_df_filtered.copy()
reviews_df_filtered_mapped['user_id'] = reviews_df_filtered_mapped['user_id'].apply(lambda x: user_mapping[x])
reviews_df_filtered_mapped['item_id'] = reviews_df_filtered_mapped['business_id'].apply(lambda x: item_mapping[x])
reviews_df_filtered_mapped.head()
```

```python
def date2timestamp(date_str: str) -> int:
    """
     
    """
    date_obj = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
    timestamp_seconds = datetime.timestamp(date_obj)
    return timestamp_seconds * 1000
```

```python
reviews_df_filtered_mapped['timestamp'] = reviews_df_filtered_mapped['date'].apply(lambda x: date2timestamp(x))
```

```python
reviews_kept_columns = ['user_id', 'item_id', 'timestamp', 'stars']
business_kept_columns = ['item_id', 'name', 'categories', 'stars']
```

```python
reviews_df_filtered_mapped[reviews_kept_columns].to_csv("yelp_interactions.csv", index=False)
business_filtered[business_kept_columns].to_csv("yelp_business.csv", index=False)
```
